```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['John', 'Alice'],
    'Age': ['25', '30'],
    'Salary': ['50000', '55000']
})
print(df.dtypes)
```

```python
df['Age'] = df['Age'].astype(int)        # To integer
df['Salary'] = df['Salary'].astype(float) # To float
print(df.dtypes)
```

```python
df['Age']= pd.to_numeric(df['Age'], errors='coerce')
print(df.dtypes)
```
